# Cosine similarity

```python
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine,correlation
```

```python
u_sim = 1-pairwise_distances(us_df.values,metric='cosine')
```

```python
u_sim
```

```python
u_s_df = pd.DataFrame(u_sim)
```

```python
u_s_df.shape
```

```python
u_s_df.index = us_df.index
u_s_df.columns = us_df.index
```

```python
def get_reco(userid):
    tem = list(u_s_df.sort_values([userid],ascending=False).head(100).index)
    book_list = list(rb1[rb1["userid"].isin(tem)]["Booktitle"])
    return set(book_list)-set(rb1[rb1["userid"]==userid]["Booktitle"])
```
